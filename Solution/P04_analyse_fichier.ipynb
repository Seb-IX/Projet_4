```python
%matplotlib inline
import pandas as pd 
import numpy as np
import os
```

```python
def decryptage_data(col_describ_data,file_name,col=None):
    descripteur = col_describ_data.loc[col_describ_data["Table"] == file_name][["Row","Description","Special"]]
    print("Description du fichier", file_name)
    for i in range(descripteur.shape[0]):
        row = descripteur.iloc[i]["Row"]
        description = descripteur.iloc[i]["Description"]
        special = descripteur.iloc[i]["Special"]
        if str(special) != "nan":
            print("- ",row,"(",special,")"," : ", description)
        else:
            print("- ",row," : ", description )

```

```python
dir_data = "./data/"
print("Liste des fichiers :")
for file in os.listdir(dir_data):
    d = pd.read_csv(dir_data + file)
    print("- ", file ," " * (35 - len(file)), "Dimensionnalité : ", d.shape )
```

```python
col_describ_data = pd.read_csv(dir_data + "HomeCredit_columns_description.csv")
print("Nombre de colonne pour chaque fichier :")
col_describ_data.Table.value_counts()
```

```python
for file in np.unique(col_describ_data.Table):
    print("\n","-"*100,"\n")
    decryptage_data(col_describ_data,file)
```
